```python
import numpy as np
import matplotlib.pyplot as plt
from PointsReducer import PointsReducer
pr = PointsReducer()
```

```python
# Generate some data
Npts = 5000
points = np.random.rand( Npts, 2).astype( dtype=np.float32 ) - 0.5
keep = np.ones( (Npts,1), dtype=np.bool )

pr.reduce_points( points, Npts, 0.05, 11, keep )


colors = np.zeros( (Npts,3), dtype=np.uint32 );
colors[ np.logical_not( keep[:,0] ),:] = (1,0,0)
plt.scatter( points[:,0], points[:,1], c=colors)
plt.show()


points_good = points[ keep[:,0], :]
plt.scatter( points_good[:,0], points_good[:,1])
plt.show()
```
