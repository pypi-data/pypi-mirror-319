```python
from tabpfn import TabPFNRegressor
import numpy as np

train_X = np.random.rand(100, 10)
train_y = np.random.rand(100)

test_X = np.random.rand(100, 10)
test_y = np.random.rand(100)

model = TabPFNRegressor(n_estimators=1)
model.fit(train_X, train_y)
```

```python
model.predict(test_X, output_type="main")["quantiles"]
```

```python
for q in model.predict(test_X, output_type="main")["quantiles"]:
    print(q.shape)
```
