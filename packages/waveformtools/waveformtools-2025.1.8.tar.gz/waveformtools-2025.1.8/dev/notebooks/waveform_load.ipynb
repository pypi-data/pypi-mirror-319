```python
import numpy as np
from waveformtools.waveforms import modes_array
import matplotlib.pyplot as plt
```

```python
sim_label = "eccprecrun4b_lev3"  # A name for the waveform
data_dir = "/mnt/pfs/vaishak.p/Projects/Codes/waveforms/eccprecrun4b_waveforms_Lev3/extrapolated"  # The path to the directory that contains the waveform
file_name = "rhOverM_Extrapolated_N2_CoM.h5"  # The name of the file containing the modes to be loaded.
```

```python
wfm = modes_array(label=sim_label, data_dir=data_dir, file_name=file_name)
```

```python
wfm.load_modes(ftype="SpEC", extrap_order=3, resam_type="coarsest")
```

```python
wfm.plot_modes(modes_to_plot=[(2, 0)])
```

The 2,0 mode is usually constant because it describes axisymmetric radiation being received from the system and in circular orbits, orbit averaged quantities are constant with time. However, in an eccentric system, the orbit averages quantities should decreate with time because eccentriity is lost.

The question is, does precession also introduce similar modulation? If yes, then is its tim period the same as that of orbit? Precession time scale can be different from ourbital time scale.
